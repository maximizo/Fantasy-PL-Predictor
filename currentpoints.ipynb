```python
import requests
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None
```

```python
#import fantasy api url with current data 
data_url = 'https://fantasy.premierleague.com/api/bootstrap-static/'
#get request from api endpoint
r = requests.get(data_url)
#transform into json object
json = r.json()
#preview json keys
json.keys()
```

```python
#create dataframe
players_df = pd.DataFrame(json['elements'])
players_df
```

```python
stats_df = pd.DataFrame(json['element_stats'])
stats_df
```

```python
gameweeks_df = pd.DataFrame(json['events'])
gameweeks_df
```

```python
positions_df = pd.DataFrame(json['element_types'])
positions_df
```

```python
teams_df = pd.DataFrame(json['teams'])
teams_df
```

```python
#check all data columns of the players
players_df.columns
```

```python
#Keep the columns we could use
cut_players_df = players_df[['id','second_name','first_name','minutes','value_season','value_form','form','total_points','points_per_game','team','team_code','in_dreamteam','now_cost','event_points','chance_of_playing_next_round','code','dreamteam_count','element_type','goals_scored','assists','clean_sheets','goals_conceded','own_goals','penalties_saved','penalties_missed','yellow_cards','red_cards','saves','bonus','bps','influence','creativity','threat','ict_index'
                    ]]
cut_players_df.head(10)
```

```python
#merge player position (abbreviation) into players file
cut_players_df['position'] = cut_players_df.element_type.map(positions_df.set_index('id').plural_name_short)
```

```python
cut_players_df.head(5)
```

```python
#merge team name to player file
cut_players_df['team'] = cut_players_df.team.map(teams_df.set_index('id').name)
```

```python
cut_players_df.head(5)
```

```python
#sort data by points
cut_players_df_sorted = cut_players_df.sort_values('total_points', ascending=False)
cut_players_df_sorted
```

```python
#Export to CSV file 
cut_players_df_sorted.to_csv('currentpoints.csv', index=False)
```

```python
#Export to CSV file 
cut_players_df_sorted.to_csv('currentpoints.zip', index=False)
```
