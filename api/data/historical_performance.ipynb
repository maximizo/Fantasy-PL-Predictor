```python
import pandas as pd
```

```python
file_17 = "resources/playerdata_201718.csv"
file_18 = "resources/playerdata_201819.csv"
file_19 = "resources/playerdata_201920.csv"
```

```python
df_17 = pd.read_csv(file_17)
df_18 = pd.read_csv(file_18)
df_19 = pd.read_csv(file_19)
```

```python
df_17.head()
```

```python
for col in df_17.columns:
    print(col)
```

```python
df_17 = df_17.drop(columns=["minutes_played_home","minutes_played_away","appearances_home", "appearances_away", "goals_home", "goals_away", "assists_home", "assists_away", "clean_sheets_home", "clean_sheets_away", "conceded_home", "conceded_away", "goals_per_90_home", "goals_per_90_away"])
df_18 = df_18.drop(columns=["minutes_played_home","minutes_played_away","appearances_home", "appearances_away", "goals_home", "goals_away", "assists_home", "assists_away", "clean_sheets_home", "clean_sheets_away", "conceded_home", "conceded_away", "goals_per_90_home", "goals_per_90_away"])
df_19 = df_19.drop(columns=["minutes_played_home","minutes_played_away","appearances_home", "appearances_away", "goals_home", "goals_away", "assists_home", "assists_away", "clean_sheets_home", "clean_sheets_away", "conceded_home", "conceded_away", "goals_per_90_home", "goals_per_90_away"])
```

```python
merge_df = df_17.merge(df_18, how="left", on=["full_name", "birthday"], suffixes=("_201718", "_201819"))
merge_df.head()
```

```python
final_df = merge_df.merge(df_19, how="left", on=["full_name", "birthday"])
final_df.head()
```

```python
final_df.columns
```

```python
updated_col_names = {}
for field in ['age', 'league', 'season', 'position', 'Current Club',
       'minutes_played_overall', 'nationality', 'appearances_overall',
       'goals_overall', 'assists_overall', 'penalty_goals', 'penalty_misses',
       'clean_sheets_overall', 'conceded_overall', 'yellow_cards_overall',
       'red_cards_overall', 'goals_involved_per_90_overall',
       'assists_per_90_overall', 'goals_per_90_overall',
       'min_per_goal_overall', 'conceded_per_90_overall',
       'min_per_conceded_overall', 'min_per_match', 'min_per_card_overall',
       'min_per_assist_overall', 'cards_per_90_overall',
       'rank_in_league_top_attackers', 'rank_in_league_top_midfielders',
       'rank_in_league_top_defenders', 'rank_in_club_top_scorer']:
    new_field = field + "_201920"
    updated_col_names[field] = new_field

updated_col_names
    
```

```python
final_df = final_df.rename(columns=updated_col_names)
final_df.head()
```

```python
for col in final_df.columns:
    print(col)
```

```python
final_df.to_csv("resources/cleaned_player_data.csv")
```
