```python
#import fantasy api url for past seasons stats per player
#player_url = 'https://fantasy.premierleague.com/api/element-summary/'+player_int+'/'
player_url = 'https://fantasy.premierleague.com/api/element-summary/4/'
```

```python
#get request from api endpoint
r = requests.get(player_url)
#transform into json object
json = r.json()
#preview json keys
json.keys()
```

```python
#create dataframes using the 3 keys
history_past_df = pd.DataFrame(json['history_past'])
fixtures_df = pd.DataFrame(json['fixtures'])
history_df = pd.DataFrame(json['history'])
```
